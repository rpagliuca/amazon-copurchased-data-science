## Load the libraries

Let's load the Python libraries that we will need throughout the script

```python
%load_ext autoreload
%autoreload 1
%aimport shared_functions
import pandas as pd
import numpy as np
from __future__ import division
import shared_functions
from IPython.display import display, HTML
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.optimize import curve_fit
from IPython.display import display, Math
```

## Experimental data

```python
data_500_estimators = [
    [500, 47],
    [5000, 154],
    [10000, 498],
    [15000, 1304]
]

data_20_estimators = [
    [5000, 24],
    [10000, 34],
    [15000, 64],
    [20000, 110],
    [25000, 197],
    [50000, 419],
    [100000,]
]

data = data_20_estimators
```

```python
x = [d[0] for d in data]
y = [d[1] for d in data]
y_lin = [d[0]/1000 for d in data]
y_exp = [np.exp(d[0]/5000) for d in data]
```

# Fit linear

```python
def func(x, b, a):
    return a*x + b

reg = curve_fit(func, x, y)

def lin_func(x):
    return func(x, *reg[0])

y_lin = [lin_func(i) for i in x]
```

## Obtained model

```python
display(Math(r'y_{{lin}}(x) = {} + {}x'.format(*[round(i*10000)/10000 for i in reg[0]])))
```

# Fit exponential

```python
def func(x, d, b, c, a):
    return a*b**(c*x) + d

reg = curve_fit(func, x, y)

def exp_func(x):
    return func(x, *reg[0])

y_exp = [exp_func(i) for i in x]
```

## Obtained model

```python
display(Math(r'y_{{exp}}(x) = {} + {} \cdot {}^{{{}x}}'.format(*[round(i*10000)/10000 for i in reg[0]])))
```

# Plot

```python
plt.figure(figsize=(8,8), dpi=130)                                                                                  
plt.plot(x, y, marker='o', label=u'Resultado experimental')
plt.plot(x, y_lin, marker='o', label=u'Fit linear')
plt.plot(x, y_exp, marker='o', label=u'Fit exponencial')
plt.ylabel(u'Tempo de execução (em segundos)')
plt.xlabel(u'Número de arestas utilizadas pela rotina de estimativa de preços')
plt.title(u'Tempo de execução vs. número de arestas\nutilizadas pela rotina de estimativa de preços')
plt.legend()
plt.show()
```
